```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('upper_torso_dict_json.csv')
df_T = df.T
df_T
```

```python
df_T.columns = ['x', 'y', 'z']
df_T
```

```python
frame = []
for i in range(241):
    frame.append(i+1)
frame_col = pd.DataFrame(frame)
frame_col.columns = ['frame']
frame_col
```

```python
df_T.index = frame_col.index
new_df = pd.concat([df_T, frame_col], axis=1)
new_df
```

```python
col1 = new_df[['x', 'frame']]
col1_df = pd.DataFrame(data=col1)
col1_df
```

```python
col2 = new_df[['y', 'frame']]
col2_df = pd.DataFrame(data=col2)
col2_df
```

```python
col3 = new_df[['z', 'frame']]
col3_df = pd.DataFrame(data=col3)
col3_df
```
